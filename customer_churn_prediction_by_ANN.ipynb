```python
dataset = dataset.drop('Surname', axis=1)
```

```python
dataset = pd.get_dummies(dataset, columns=['Geography'], drop_first=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset['Gender'] = le.fit_transform(dataset['Gender'])
```

```python
print(dataset.dtypes)
```

```python
input_data=dataset.iloc[:,:-1]
output_data=dataset.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
from re import X
X_train,X_test,y_train,y_test=train_test_split(input_data,output_data,test_size=0.2,random_state=42)
```

```python
from keras.layers import Dense
from keras.models import Sequential

ann=Sequential()

# Change the input_dim to 13 to match the number of features in your data
ann.add(Dense(10, input_dim=13, activation="relu"))
ann.add(Dense(8, activation="relu")) # This layer expects 10 features as input from previous layer
ann.add(Dense(2, activation="relu")) # This layer expects 8 features as input from previous layer
ann.add(Dense(1, activation="sigmoid")) # This layer expects 2 features as input from previous layer

ann.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])

ann.fit(X_train,y_train,batch_size=100,epochs=50)
```

```python
prd1=ann.predict(X_train)
prd_data1=[]
for i in prd1:
  if i[0]>0.5:
    prd_data1.append(1)
  else:
    prd_data1.append(0)
```
